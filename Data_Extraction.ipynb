```python
import pandas as pd
from sklearn.model_selection import train_test_split, 
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
data_urlK = "https://exoplanetarchive.ipac.caltech.edu/cgi-bin/nstedAPI/nph-nstedAPI?table=cumulative&select=kepid,kepoi_name,koi_disposition,koi_period,koi_time0bk,koi_impact,koi_duration,koi_depth,koi_prad,koi_sma,koi_teq,koi_insol,koi_model_snr,koi_steff,koi_slogg,koi_srad,koi_smass,ra,dec,koi_kepmag&format=csv"
data_urlT = "https://exoplanetarchive.ipac.caltech.edu/cgi-bin/nstedAPI/nph-nstedAPI?table=toi&select=toi,toipfx&format=csv"
df_kepler = pd.read_csv(data_urlK)
df_kepler.head()
```

```python
df_kepler_clean=df_kepler.dropna().reset_index(drop=True)
df_kepler_clean
```

```python
df=df_kepler_clean.drop(['kepid', 'kepoi_name'], axis=1)
df
```

We take only information about the classes CONFIRMED and FALSE POSITIVE for the binary classification in order to obtain a model that can be used on the CANDIDATE class.

```python
df=df[df['koi_disposition']!='CANDIDATE'].reset_index(drop=True)
df
```

```python
df['koi_disposition'].unique()
```

```python
le=LabelEncoder()
df['koi_disposition']=le.fit_transform(df['koi_disposition'])
df
```

```python
X=df.drop('koi_disposition',axis=1).values
y=df['koi_disposition'].values
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
X_train
```

```python
scaler=StandardScaler()
X_train_scaled_1=scaler.fit_transform(X_train)
X_test_scaled_1=scaler.transform(X_test)
X_train_scaled_1
X_test_scaled_1
```
